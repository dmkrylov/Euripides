```python
import unicodedata

with open ('C:\\Users\\Arseny\\Desktop\\aeschylus_1.txt', 'r', encoding = 'utf-8') as agamemnon_1:
    agamemnon_1 = agamemnon_1.read()
stichoi = agamemnon_1.split('\n')
strings =[]
for stichos in stichoi:
    stichos = stichos.lower()
    strings.append(stichos)
    
yihhi = 'ὕἶῃὴΐ'
xyhta = unicodedata.normalize('NFD',yihhi)
diacritics = xyhta[1]+xyhta[4]+xyhta[9]+xyhta[2]+xyhta[5]+xyhta[7]+xyhta[11]
vowels = 'ωοηεαυι'
dyphtongs = ['οι','ει','αι','υι','ου','ευ','αυ','ηυ']

def syllabization (s):
    syllables = 0
    s = unicodedata.normalize('NFD', s)
    for symbol in s:
        if symbol in vowels:
            syllables +=1
    for i in range (len(s)-1):
        if s[i]+s[i+1] in dyphtongs:
            syllables -=1
    for j in range (len(s)-2):
        if s[j]+s[j+1] in dyphtongs and s[j+2]==diacritics[-1]:
            syllables +=1
    if syllables != 12:
        print (s)
    else:
        print ((syllables))

for stichos in strings:
    syllabization (stichos)
```
